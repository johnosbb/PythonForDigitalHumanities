```python
from bs4 import BeautifulSoup
import requests

url = "https://en.wikipedia.org/wiki/Nick_Drake"
file ="./ReferenceFiles/wikipedia.txt"

s = requests.get(url)
soup = BeautifulSoup(s.text, "lxml") #   read page as html




def contents_extract():
    contents = soup.find_all("p") # get all p tags
    with open(file,"a+") as f:
        for p in contents:
            f.write(p.get_text() + "\n")
            
def sources_extract():
    sources = soup.find_all("cite") # get all citations
    with open(file,"a+") as f:
        for s in sources:
            f.write(s.get_text() + "\n")
        
        
def heading_extract():
    heading = soup.find("h1",{"id" : "firstHeading"})
    print(heading.get_text())
    with open(file,"a+") as f:
        f.write(heading.get_text())
        
def table_extract():
    table = soup.find("table",{"class" : "infobox vcard plainlist"})
    with open(file,"a+") as f:
        for row in table.find_all("tr"): # get all of the rows
            for th in row.find_all("th"): # get the headings
                f.write("\n" + th.get_text() + ": ")      
                for td in row.find_all("td"): # get the cells in each row
                    f.write(" " + td.get_text() + " ") 
        
heading_extract()
contents_extract()
sources_extract()  
table_extract()      
```
