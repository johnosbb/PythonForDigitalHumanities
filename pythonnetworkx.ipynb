```python
import networkx as nx
import matplotlib.pyplot as plt


g = nx.Graph() # create a graph
g.add_node("Tom")
g.add_node("Jerry")
g.add_node("Susan")

g2 = nx.Graph() # create a graph
g2.add_node("Freddy")
g2.add_node("Lucy")
g2.add_node("John")


print(list(g))
for i in g:
    print(i)
print(nx.nodes(g))   
print(len(g)) 
print(nx.number_of_nodes(g))


plt.figure(1)
nx.draw(g,node_color="green")
plt.figure(2)
nx.draw(g2,node_color="red")
plt.show()



```

```python
import networkx as nx
import matplotlib.pyplot as plt



list_names = [("Tom","Jerry"),("Fred","Wilma"),("Mike","Rita"),("June","John"),("Tom","Sam"),("Rita","Fred"),("Mike","Fred")]
g = nx.Graph() # create a graph
g.add_node("Tom")
g.add_node("Jerry")
g.add_node("Susan")
#g.add_edge("Tom","Jerry")
g.add_edges_from(list_names)
g.add_edge("Jerry","Sam")
g.add_edge("Rita","June")
g.add_edge("Jerry","John")
#g.remove_edge("Fred","Wilma")
print(nx.has_path(g,"Mike","Rita"))
print(nx.shortest_path(g,"Mike","June"))
print(nx.single_source_shortest_path(g,"Mike",2))
plt.figure(1)
nx.draw(g,node_size= 1000,node_color="green",with_labels=True,font_size = 10,font_color = "White",bbox=dict(facecolor="Blue",alpha = .5) )

plt.show()
```
