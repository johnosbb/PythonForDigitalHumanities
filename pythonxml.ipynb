```python
import xml.etree.ElementTree as ET

tree = ET.parse('./ReferenceFiles/bookstore.xml')
root = tree.getroot()

# Iterate over each <book> element
for book in root.findall('book'):
    # Get the value of the category attribute
    category = book.get('category')
    print(f"Category: {category}")
    
    # Get the values of the child elements
    title = book.find('title').text
    author = book.find('author').text
    year = book.find('year').text
    
    print(f"Title: {title}")
    print(f"Author: {author}")
    print(f"Year: {year}")
    print()
```

```python
import xml.etree.ElementTree as ET

tree = ET.parse('./ReferenceFiles/rc.xml')
root = tree.getroot()

#keyboard = root.find('{http://openbox.org/3.4/rc}keyboard')
keyboard = root.find('rc:keyboard', {'rc': 'http://openbox.org/3.4/rc'})
if keyboard:
    #print(keyboard)
    for keybind in keyboard:
        
        if(keybind.tag == '{http://openbox.org/3.4/rc}keybind'):
            for action in keybind:
                print(f" action : {action.tag} {action.text}")
                attributes = action.attrib
                if attributes:
                    print("Action Attributes:")
                for attr_name, attr_value in attributes.items():
                    print(f"{attr_name}: {attr_value}")
            print(f"Keybind tag {keybind.tag}, text: {keybind.text}")
            attributes = keybind.attrib
            if attributes:
                print("Attributes:")
                for attr_name, attr_value in attributes.items():
                    print(f"{attr_name}: {attr_value}")
```
